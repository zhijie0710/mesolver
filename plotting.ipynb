```python
import matplotlib.pyplot as pt
import numpy as np
```

```python
t, f1, f2 = np.loadtxt("qubitResonator_correct_output.dat")
pt.figure(figsize=(5,3))
s1, = pt.plot(t, f1, label=r"$|00\rangle$")
s2, = pt.plot(t, f2, label=r"$|11\rangle$")
pt.legend(handles=[s1, s2], loc=5)
pt.xlabel("time")
pt.ylabel("expectation value")
pt.savefig("qubitResonator_correct_output.png", dpi=300)
pt.show()
```
